### Data Cleaning

```python
# numeric column with missing values -> handle it with median
ca_median = df['ca'].median()
print(ca_median)
```

```python
df['ca'].fillna(ca_median, inplace=True)
```

```python
print(df['ca'].isna().sum())
```

```python
# categorical column with missing values -> handle it with mode
thal_mode = df['thal'].mode()[0]
print(thal_mode)
```

```python
df['thal'].fillna(thal_mode, inplace=True)
```

```python
df['thal'].isna().sum()
```

```python
df.head(8)
```

```python
# convert target variable to binary classification
df['num'] = df['num'].apply(lambda x: 1 if x > 0 else 0)
```
